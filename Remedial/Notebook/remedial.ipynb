```python
# Ruta al archivo CSV en tu directorio actual
ruta_archivo = 'marketing_campaign.csv'

# Cargar el dataset desde la ruta especificada
df = pd.read_csv(ruta_archivo, sep=';')

# Mostrar las primeras filas del dataframe para verificar la carga
df.head()
```

```python
# Limpiar NaNs (eliminar filas con valores faltantes)
df_cleaned = df.dropna()
```

```python
# Limpiar NaNs (rellenar con un valor específico, como 0)
df_cleaned = df.fillna(0)
```

```python
# Reindexar el dataset
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
# Renombrar columnas para mayor claridad
df_cleaned.rename(columns={'old_name1': 'new_name1', 'old_name2': 'new_name2'}, inplace=True)
```

```python
import numpy as np

# Seleccionar solo columnas numéricas
df_numeric = df_cleaned.select_dtypes(include=[np.number])

# Media
mean = df_numeric.mean()

# Mediana
median = df_numeric.median()

# Media truncada (elimina el 10% superior e inferior de los datos y calcula la media)
def trimmed_mean(series, proportion_to_cut=0.1):
    sorted_series = series.sort_values()
    n = len(sorted_series)
    cut = int(n * proportion_to_cut)
    return sorted_series[cut:-cut].mean()

trimmed_mean_values = df_numeric.apply(trimmed_mean)

# Rango (diferencia entre el valor máximo y mínimo)
range_val = df_numeric.max() - df_numeric.min()

# Valor mínimo
min_val = df_numeric.min()

# Valor máximo
max_val = df_numeric.max()

mean, median, trimmed_mean_values, range_val, min_val, max_val
```

```python
import numpy as np

# Seleccionar solo columnas numéricas
df_numeric = df_cleaned.select_dtypes(include=[np.number])

# Verificar el tipo de datos de las columnas seleccionadas
print(df_numeric.dtypes)

# Forzar conversión a numérico (con errores convertidos a NaN)
df_numeric = df_cleaned.apply(pd.to_numeric, errors='coerce')

# Volver a filtrar las columnas numéricas después de la conversión
df_numeric = df_numeric.select_dtypes(include=[np.number])

# Media
mean = df_numeric.mean()

# Mediana
median = df_numeric.median()

# Media truncada (elimina el 10% superior e inferior de los datos y calcula la media)
def trimmed_mean(series, proportion_to_cut=0.1):
    sorted_series = series.sort_values()
    n = len(sorted_series)
    cut = int(n * proportion_to_cut)
    return sorted_series[cut:-cut].mean()

trimmed_mean_values = df_numeric.apply(trimmed_mean)

# Rango (diferencia entre el valor máximo y mínimo)
range_val = df_numeric.max() - df_numeric.min()

# Valor mínimo
min_val = df_numeric.min()

# Valor máximo
max_val = df_numeric.max()

mean, median, trimmed_mean_values, range_val, min_val, max_val
```
